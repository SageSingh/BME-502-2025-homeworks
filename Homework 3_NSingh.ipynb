### 1.1 

#### What is the chance of the knight crossing the bridge safely?


```julia
# Law of Total Probability: P(A) = P(A|B_n)(P(B_n))+P(A|B_n+1)(P(B_n+1)).....
# P(Gnome) = P(Gnome|E1)P(E1)+P(Gnome|E2)P(E2)+P(Gnome|E3)P(E3)

## Vars 
probE1=.25
probE2=.25
probE3=.50

Prob_GgivenE1 = .6 # 3/5 chance of good gnomes to be interacted with. 
Prob_GgivenE2 = .8 # 4/5 chance of good gnomes to be interacted with. 
Prob_GgivenE3 = 1.0 # 5/5 chance of good gnomes to be interacted with. 

# Law of Total Prob to Solve P(G)
prob_GnomeCase1=(Prob_GgivenE1*probE1) +(Prob_GgivenE2*probE2) + (Prob_GgivenE3*probE3)
println("Safety Probability = ", prob_GnomeCase1*100 ,"% ")
```

### 1.3

#### Now he wants to use Bayes' theorem to calculate his chance of crossing safely given a troll captured 

\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posteior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

#### Given the three events, E1, E2, and E3, which two of the three events does the knight need to consider after a troll was captured?

```julia
# Law of Total Probability: P(A) = P(A|B_n)*P(B_n) + P(A|B_n+1)*P(B_n+1) + ...
# Example: P(Troll) = P(T|E1)*P(E1) + P(T|E2)*P(E2) + P(T|E3)*P(E3)

# Vars
probE1 = 0.25
probE2 = 0.25
probE3 = 0.50

# Modified likelihoods after capturing/removing a troll, etc.
# E1 = TTGGG [originally 2/5 trolls] → after removal, TGGG (1 troll, 3 good) => 3/4 chance good
# E2 = TGGGG [originally 1/5 trolls] → after removal, GGGG (0 trolls, 4 good) => 4/4 chance good
# E3 = GGGGG [originally 0/5 trolls] → still 0 trolls => 5/5 chance good

Prob_GgivenE1 = 0.75  # 3/4
Prob_GgivenE2 = 1.0   # 4/4
Prob_GgivenE3 = 1.0   # 5/5

# Law of Total Probability for P(Good) in the updated scenario
prob_GnomeCase2 = (Prob_GgivenE1 * probE1) +
                  (Prob_GgivenE2 * probE2) +
                  (Prob_GgivenE3 * probE3)

# Probability of danger (i.e., encountering a troll) is 1 - P(Good)
prob_danger = 1 - prob_GnomeCase2

# Example "previous case" probability (if defined previously)
prob_GnomeCase1 = 0.55  # or some other value from a prior calculation

println("Safety Probability (Case 2) = $(prob_GnomeCase2 * 100)%")

println("The risk from Case E1 is that there's still 1 troll out of 4 creatures, ",
        "with a danger risk of $(prob_danger * 100)%")
```

### 2.3

#### Fit the data (x, y_observed) with a second order polynomial. 

1. Make a scatter plot of y_observed along x
2. Plot a line of y_true along x on the scatter plot
3. Plot a line of the fitted polynomial along x on the same plot

```julia
using Plots, Random, Distributions
import Distributions as di # importing Distributions pkg, but shorthanding this now as ".di" appending
using Polynomials: Polynomial, fit, coeffs # From Polynomial Pkg, ONLY import these 3 functions "Polynomial,fir, Coeffs"
using Statistics 

# Vars and Eqns
x = -10:10
y_true = x.^2 .+ 4 .* x .- 2 # Second order polynom.
nd = di.Normal(0, 10) # mu = 0, sd = 10

#Observed with random "noise" data
y_observed = y_true .+ rand(nd, length(x))

#Plot the "true" data and the "observed" (noisy) data
plot(x,y_true,label= "True")
plot!(x,y_observed,label="Observed")

poly_fit = fit(x, y_observed, 2)
x_fit= -10:1:10 # same as X
y_polyfit = poly_fit.(x_fit)

#Plots with X and Y fit modified
scatter(x,y_true,label="True")
plot!(x,y_observed,label="Observed")
plot!(x_fit,y_polyfit,label="Fitted")
```
